```python
import matplotlib.pyplot as plt
import numpy as np
```

### 1) Line Plot with only y input

```python
y = np.array([2, 4, 6, 5, 3, 7, 1, 5])
```

```python
plt.plot(y)
plt.xlabel("x input")
plt.ylabel("y input")
plt.title("Providing only y input")
plt.show()
```

### 2) Line Plot with both x and y inputs

```python
x = np.array([0, 1, 2, 3, 4, 5, 6, 7])
y = np.array([2, 4, 6, 5, 3, 7, 1, 5])
```

```python
plt.plot(x, y)
plt.xlabel("x input")
plt.ylabel("y input")
plt.title("Providing both x and y inputs")
plt.show()
```

### 3) Choosing line color

```python
plt.plot(x, y, color='red')
plt.xlabel("x input")
plt.ylabel("y input")
plt.title("Line color: red")
plt.show()
```

### 4) Changing line width

```python
plt.plot(x, y, color='red', linewidth=3)
plt.xlabel("x input")
plt.ylabel("y input")
plt.title("Changing line width")
plt.show()
```

### 5) With marker on the line

```python
plt.plot(x, y, marker='o', color='red')
plt.xlabel("x input")
plt.ylabel("y input")
plt.title("circle marker")
plt.show()
```

```python
plt.plot(x, y, color='red', marker='*')
plt.xlabel("x input")
plt.ylabel("y input")
plt.title("star marker")
plt.show()
```

```python
plt.plot(x, y, color='red', marker='D')
plt.xlabel("x input")
plt.ylabel("y input")
plt.title("diamond marker")
plt.show()
```

### 6) Changing marker size

```python
plt.plot(x, y, color='red', marker='D', markersize=10)
plt.xlabel("x input")
plt.ylabel("y input")
plt.title("Changing marker size")
plt.show()
```

### 7) Different color for line and marker

```python
plt.plot(x, y, marker='D', markersize=10, markerfacecolor='pink', markeredgewidth=2, markeredgecolor='blue', color='red')
plt.xlabel("x input")
plt.ylabel("y input")
plt.title("Different color for line and marker")
plt.show()
```

### 8) Changing line style

```python
plt.plot(x, y, color='deeppink', linestyle='dashed', marker='o')
plt.xlabel("x input")
plt.ylabel("y input")
plt.title("Dashed lines")
plt.show()
```

```python
plt.plot(x, y, color='deeppink', linestyle='dotted', marker='o')
plt.xlabel("x input")
plt.ylabel("y input")
plt.title("Dotted lines")
plt.show()
```

```python
plt.plot(x, y, color='deeppink', linestyle='dashdot', marker='o')
plt.xlabel("x input")
plt.ylabel("y input")
plt.title("Dash-dotted lines")
plt.show()
```

### 9) Using format string for easy plotting

```python
plt.plot(x, y, 'bo--')
plt.xlabel("x input")
plt.ylabel("y input")
plt.title("Using format string to plot blue dashed lines with circle maker")
plt.show()
```

### 10) How to plot multiple lines in same plot and name them

```python
# Line 1
x1 = np.array([1, 3, 5, 7, 9])
y1 = np.array([2, 6, 4, 8, 1])
# Line 2
x2 = np.array([0, 2, 4, 6, 8])
y2 = np.array([5, 2, 6, 8, 6])
```

```python
# Plot lines in default colors
plt.plot(x1, y1, label="Line 1")
plt.plot(x2, y2, label="Line 2")
plt.xlabel("x inputs")
plt.ylabel("y inputs")
plt.title("Two lines in same plot - default color")
plt.legend() # This is used to name lines in plot
plt.show()
```

```python
# Plot lines using custom colors
plt.plot(x1, y1, color='magenta', label="Line 1")
plt.plot(x2, y2, color='limegreen', label="Line 2")
plt.xlabel("x inputs")
plt.ylabel("y inputs")
plt.title("Two lines in same plot - custom color")
plt.legend() # This is used to name lines in plot
plt.show()
```
