# Matplotlib - Pie Chart

Please visit https://mldoodles.com/matplotlib-pie-chart/ for detailed explanation to this notebook 

```python
import matplotlib.pyplot as plt
```

```python
yogurts_sold = [300, 250, 160, 230, 180]
flavors = ['chocolate', 'strawberry', 'honeybanana', 'blueberry', 'pistachio']
```

### 1) Simple pie chart

```python
plt.pie(yogurts_sold, labels=flavors)
plt.show()
```

### 2) Make a pie chart bigger

```python
plt.pie(yogurts_sold, labels=flavors, radius=1.5) # default radius is 1
plt.show()
```

### 3) Using explode in pie chart

```python
explode = [0.1, 0, 0, 0, 0]
plt.pie(yogurts_sold, explode=explode, labels=flavors, radius=1.5)
plt.show()
```

### 4) Pie chart with shadow

```python
plt.pie(yogurts_sold, explode=explode, labels=flavors, radius=1.5, shadow=True)
plt.show()
```

### 5) Pie chart with different start angle 

```python
plt.pie(yogurts_sold, explode=explode, labels=flavors, radius=1.5, shadow=True, startangle=90)
plt.show()
```

### 6) Rotate labels to the angle of the wedge

```python
plt.pie(yogurts_sold, explode=explode, labels=flavors, radius=1.5, shadow=True, rotatelabels=True)
plt.show()
```

### 7) Customize colors in pie chart

```python
colors = ['indianred', 'deeppink', 'gold', 'mediumorchid', 'palegreen']
plt.pie(yogurts_sold, explode=explode, labels=flavors, colors=colors, radius=1.5, shadow=True)
plt.show()
```

### 8) Label a pie chart with percent

```python
plt.pie(yogurts_sold, explode=explode, labels=flavors, colors=colors, radius=1.5, shadow=True, autopct='%.1f%%')
plt.show()
```

### 9) Label a pie chart with percent and value

```python
plt.pie(yogurts_sold, explode=explode, labels=flavors, colors=colors, radius=2, shadow=True,  
        autopct=lambda p: "{:.1f}% ({:.0f} units)".format(p, p*sum(yogurts_sold)/100))
plt.show()
```

### 10) White and bold font for text in pie chart

```python
textprops={"color":"white",
           "weight":"bold"}     

plt.pie(yogurts_sold, explode=explode, labels=flavors, colors=colors, radius=2, shadow=True, 
        autopct=lambda p: "{:.1f}% ({:.0f} units)".format(p, p*sum(yogurts_sold)/100),
        textprops=textprops)
plt.show()
```

### 11) Pie chart with legend

```python
plt.figure(figsize=(10,8))
plt.pie(yogurts_sold, explode=explode, labels=flavors, colors=colors, shadow=True, 
       autopct= lambda p: "{:.1f}% ({:.0f} units)".format(p, p*sum(yogurts_sold)/100),
       textprops=textprops)
plt.legend(title="Yogurt Flavors", loc="upper right")
# loc: upper left, upper right, lower left, lower right, upper center, lower center, center left, center right, center, best
plt.show()
```

### 12) Wedges with thicker line and different edge color

```python
wedgeprops = {"linewidth":3, "edgecolor":"black"}
plt.figure(figsize=(10,8))
plt.pie(yogurts_sold, explode=explode, labels=flavors, colors=colors, shadow=True, 
        autopct=lambda p: "{:.2f}% ({:.0f} units)".format(p, p*sum(yogurts_sold)/100),
        wedgeprops=wedgeprops)
plt.legend(title="Yogurt Flavors")
plt.show()
```

### 13) Without label names

```python
plt.figure(figsize=(10,8))
plt.pie(yogurts_sold, explode=explode, labels=flavors, colors=colors, shadow=True, 
        autopct=lambda p: "{:.2f}% ({:.0f} units)".format(p, p*sum(yogurts_sold)/100),
        labeldistance=None)
plt.legend(title="Yogurt Flavors")
plt.show()
```

### 14) Pie chart like a donut (Hole in the middle of a pie chart)

```python
plt.pie(yogurts_sold, labels=flavors, radius=1, colors=colors, wedgeprops={"width":0.5})
plt.show()
```
