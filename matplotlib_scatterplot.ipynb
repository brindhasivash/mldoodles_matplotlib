```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(123)
length = np.random.randint(1, 10, 20)
width = np.random.randint(1, 10, 20)
```

### Simple scatterplot

```python
plt.scatter(length, width)
plt.title("Simple scatterplot")
plt.xlabel("Length")
plt.ylabel("Width")
plt.show()
```

### Scatterplot with markers of different sizes

Next, we will plot the points in varying sizes according to the third parameter - area.
Before attempting to plot the points in varying sizes, let's see what is the default size of points used by matplotlib.

```python
# Default size of the points as mentioned in the documentation
plt.rcParams["lines.markersize"]**2
```

```python
# Plot the points with varying sizes according to the area
area = [length[i] * width[i] for i in range(len(length))]
print(area)
```

The default size is 36. But we have very tiny values for area like 7, 2, 4, 8, ... which are less than the default size 36. This will make the points look very tiny. So we use a custom size calculated from this area parameter. Just to make the points look bigger.

```python
# Plotting in custom size
# We use the calculated value of area to set the size of the points
size = [20*i for i in area]
print(size)
```

```python
plt.scatter(length, width, s=size)
plt.xlabel("Length")
plt.ylabel("Width")
plt.title("Scatterplot with markers of different sizes")
```

### Scatterplot with markers of different colors

```python
# We set 20 random values for color, as we have 20 points
np.random.seed(123)
color = np.random.randn(20)
```

```python
plt.scatter(length, width, s=size, c=color)
plt.xlabel("Length")
plt.ylabel("Width")
plt.title("Scatterplot with markers of different colors")
```

### Setting transparency of plot

```python
transparency = 0.5 # value between 0 and 1. 0 is completely transparent. 1 is solid color.
plt.scatter(length, width, s=size, c=color, alpha=transparency)
plt.xlabel("Length")
plt.ylabel("Width")
plt.title("Scatterplot - transparent")
```

### Scatterplot with different shapes of markers

```python
# List of markers available in matplotlib is given in the below link
# URL : https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers

marker = '^'
plt.scatter(length, width, s=size, c=color, alpha=transparency, marker=marker)
plt.xlabel("Length")
plt.ylabel("Width")
plt.title("Scatterplot - Different shapes of markers")
```

```python
marker = 'P'
plt.scatter(length, width, s=size, c=color, alpha=transparency, marker=marker)
plt.xlabel("Length")
plt.ylabel("Width")
plt.title("Scatterplot - Different shapes of markers")
```

### Scatterplot with edgecolors

```python
# # We set 20 random values for edge color, as we have 20 points.
# The second parameter is either 3 or 4. It corresponds to (RGB) or (RGBA). I used 3.
edgecolors = np.random.random((20, 3))
```

```python
plt.scatter(length, width, s=size, c=color, edgecolors=edgecolors)
plt.xlabel("Length")
plt.ylabel("Width")
plt.title("Scatterplot with edgecolors for markers")
```

### Adjusting edge thickness

```python
# To make the edges thick
# linewidths default value is 1.5 as per documentation
linewidths = 3
plt.scatter(length, width, s=size, c=color, linewidths=linewidths, edgecolors=edgecolors)
plt.xlabel("Length")
plt.ylabel("Width")
plt.title("Scatterplot with thick edges")
```

### Scatterplot as hollow circles

The above plot has both inner color and outer edge color. We can manipulate the inner color to be white. So that we will be left with outer edge circles only. See how we do this below.

```python
plt.scatter(length, width, s=size, c='white', linewidths=linewidths, edgecolors=edgecolors)
plt.xlabel("Length")
plt.ylabel("Width")
plt.title("Scatterplot as hollow circles")
```

### Setting colormap (color palette) - the cmap parameter

So far, we have used the default colormap. (We have not set any value for cmap parameter till now.) The default cmap is "viridis". There are lots of colormap options available. You can choose your color palette as you like.

```python
# Default colormap
plt.rcParams["image.cmap"]
```

```python
# 164 colormaps available
plt.colormaps()
```

```python
# Choosing a different colormap called "winter"
plt.scatter(length, width, s=size, c=color, cmap='winter')
plt.xlabel("Length")
plt.ylabel("Width")
plt.title("Choosing a different colormap: cmap used - winter")
```

```python
# Choosing a different colormap called "Spectral"
plt.scatter(length, width, s=size, c=color, cmap='Spectral')
plt.xlabel("Length")
plt.ylabel("Width")
plt.title("Choosing a different colormap: cmap used - Spectral")
```

```python
# Choosing a different colormap called "seismic"
plt.scatter(length, width, s=size, c=color, cmap='seismic')
plt.xlabel("Length")
plt.ylabel("Width")
plt.title("Choosing a different colormap: cmap used - seismic")
```
